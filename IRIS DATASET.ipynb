## **10. Pair Plot:**


`A Pair Plot is a type of data visualization used to display relationships between multiple variables in a dataset. It is commonly used in exploratory data analysis to understand how different variables correlate with each other.`

```python
sns.pairplot(iris)
```

The plot shown is a **pairplot** (also known as a scatterplot matrix), which is commonly used for **exploratory data analysis (EDA)** in machine learning and data science. Here are its key use cases:

1. **Understanding Relationships Between Variables**  
   - It helps visualize the relationships between different numerical features in a dataset.
   - Scatterplots show how two variables are correlated.

2. **Detecting Patterns and Trends**  
   - Linear or non-linear relationships between features can be identified.
   - Clusters or separations among data points might indicate different groups or classes.

3. **Identifying Outliers**  
   - Outliers appear as isolated points in scatter plots.

4. **Checking Multicollinearity**  
   - If two variables have a strong correlation, they might provide redundant information in predictive modeling.

5. **Feature Selection for Machine Learning Models**  
   - Helps in selecting the most relevant features for modeling by analyzing correlations.

## **11. Heat map**
`A heatmap is a graphical representation of data where individual values are represented by color intensity. It is commonly used to visualize correlations between numerical features in a dataset.`

📊 **Use Case:**  
- Helps identify highly correlated variables.
- Useful for feature selection in machine learning.  
- Detects redundant information in datasets. 🚀

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)

fig = plt.gcf()
fig.set_size_inches(10, 7)

sns.heatmap(df.corr(), annot=True, cmap='cubehelix',
            linewidths=1, linecolor='k', square=True, mask=False,
            vmin=-1, vmax=1, cbar_kws={"orientation": "vertical"}, cbar=True)

plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

## **Breaking it Down:**
`iris.data`
- This is a NumPy array containing the numerical values of the dataset.
It represents the features (sepal length, sepal width, petal length, petal width) of the Iris flowers.

`iris.feature_names`
- This is a list of column names corresponding to the features in iris.data.

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)

fig = plt.gcf()
fig.set_size_inches(10, 7)

sns.heatmap(df.corr(), annot=True, cmap='cubehelix',
            linewidths=1, linecolor='k', square=True, mask=False,
            vmin=-1, vmax=1, cbar_kws={"orientation": "vertical"}, cbar=True)

plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

### **Understanding the Correlation Heatmap of the Iris Dataset**

This heatmap represents the correlation matrix of the **Iris dataset's numerical features** (sepal length, sepal width, petal length, petal width). Each cell in the matrix shows the correlation coefficient between two features.

---

### **Key Observations:**
1. **Strong Positive Correlations:**
   - **Petal Length & Petal Width (0.96)** → Strongest correlation.
   - **Sepal Length & Petal Length (0.87)**
   - **Sepal Length & Petal Width (0.82)**
   - **Interpretation:** As petal length increases, petal width also increases significantly.

2. **Negative Correlations:**
   - **Sepal Width & Petal Length (-0.43)**
   - **Sepal Width & Petal Width (-0.37)**
   - **Interpretation:** As sepal width increases, petal length and petal width tend to decrease.

3. **Low Correlation:**
   - **Sepal Length & Sepal Width (-0.12)**  
   - **Interpretation:** Sepal length and sepal width have almost no linear relationship.
---

### **32.Area Plot:**

`An Area Plot is used to represent quantitative data over time or categories. It is similar to a line plot, but the area under the line is filled with color to emphasize magnitude.`

```python
iris.columns
```

```python
# Convert 'SepalLengthCm' back to numeric
iris['SepalLengthCm'] = pd.to_numeric(iris['SepalLengthCm'], errors='coerce')

# Area plot for multiple columns
iris.plot.area(y=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'], 
               alpha=0.4, figsize=(12, 6))

plt.show()
```

### 🔥 **Why does this work?**
1. **Ensures Numerical Data** – The `.astype('category')` conversion is removed, as area plots require numbers.
2. **Handles Errors** – `pd.to_numeric(..., errors='coerce')` ensures any non-numeric values become `NaN`, preventing further issues.
3. **Maintains Original Structure** – The area plot works correctly with the specified columns.
